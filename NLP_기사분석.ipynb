# 2-1 명사만 선택

```python
import konlpy.tag

def Kr_preprocessing_noun(text, customized_stopwords=None):
    filtered_content = re.sub(r'[^\s\d\w]','',text)
    twitter = konlpy.tag.Twitter()
    twitter_morphs = twitter.pos(filtered_content)
    Noun_words = []
    for word, pos in twitter_morphs:
        if pos == 'Noun':
            Noun_words.append(word)
    unique_Noun_words = set(Noun_words)
    for word in unique_Noun_words:
        if word in customized_stopwords or len(word)==1:
            while word in Noun_words: Noun_words.remove(word)
    return Noun_words
```

```python
stopwords = []
docs_nouns = [Kr_preprocessing_noun(doc,stopwords) for doc in total_docs]
```

```python
print(docs_nouns[0]) # 첫번째 문서에 대한 내용 ((문서 내용, [명사단어])로 구성)
```

```python
from collections import Counter

for i in range(20) :
  c = Counter(docs_nouns[i])
  print(c)
```

```python
documents_filtered_noun = []
for doc in docs_nouns:
    document_filtered_noun =' '.join(doc)
    documents_filtered_noun.append(document_filtered_noun)
documents_filtered_noun
```

# 2-2 TF 정보를 이용하여 벡터화하기

```python
from sklearn.feature_extraction.text import CountVectorizer # frequency based DTM
def tf_extractor(corpus): 
    # returns a frequency-based DTM
    vectorizer = CountVectorizer(min_df=1, ngram_range=(1,1))
    features = vectorizer.fit_transform(corpus) # transform texts to a frequency matrix
    return vectorizer, features 

vectorizer_tf, DTM_tf_noun = tf_extractor(documents_filtered_noun) # 문서들을 TF matrix로 변환 
DTM_TF_noun = np.array(DTM_tf_noun.todense())
```

```python
DTM_TF_noun
```

**실루엣 스코어를 이용한 최적의 군집 수 구하기**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
for k in range(2,10):
    kmeans = KMeans(n_clusters=k, max_iter=10, random_state = 40)
    cluster = kmeans.fit_predict(DTM_TF_noun) 
    print(k, silhouette_score(DTM_TF_noun, cluster))
```

```python
#silhouette score가 가장 높은 k : 4
#따라서 최적의 군집 수는 4개
kmeans = KMeans(n_clusters=4,max_iter=10, random_state = 40)
clusters_TF_noun = kmeans.fit_predict(DTM_TF_noun)
print(clusters_TF_noun)
```

```python
import pandas as pd
```

```python
pd.DataFrame({'clustering label' : kmeans.fit_predict(DTM_TF_noun), 'corpus' : documents_filtered_noun})
```

# 2-3 TF-IDF 정보를 이용하여 벡터화하기

```python
from sklearn.feature_extraction.text import TfidfVectorizer # tf-idf based DTM

def tfidf_extractor(corpus):
    # returns a tf-idf based DTM
    vectorizer = TfidfVectorizer()
    features = vectorizer.fit_transform(corpus)
    return vectorizer, features
    #features에 DTM 정보(vectorazation 된 정보)가 저장되어 있음

vectorizer_tfidf, DTM_tfidf_noun = tfidf_extractor(documents_filtered_noun) 
DTM_TFIDF_noun = np.array(DTM_tfidf_noun.todense())
```

```python
for k in range(2,10):
    kmeans = KMeans(n_clusters=k, max_iter=10, random_state = 40)
    cluster = kmeans.fit_predict(DTM_TFIDF_noun) 
    print(k, silhouette_score(DTM_TFIDF_noun, cluster))
```

```python
#silhouette score가 가장 높은 k : 4
#따라서 최적의 군집 수는 4개
kmeans = KMeans(n_clusters=4, max_iter=10, random_state = 40)
print(kmeans.fit_predict(DTM_TFIDF_noun))
```

# 3.1 명사와 동사 선택

```python
def Kr_preprocessing_noun_verb(text, customized_stopwords=None):
    filtered_content = re.sub(r'[^\s\d\w]','',text)
    twitter = konlpy.tag.Twitter()
    twitter_morphs = twitter.pos(filtered_content)
    Noun_Verb_words = []
    for word, pos in twitter_morphs:
        if pos == 'Noun' or pos == 'Verb' :
            Noun_Verb_words.append(word)
    unique_Noun_Verb_words = set(Noun_Verb_words)
    for word in unique_Noun_Verb_words:
        if word in customized_stopwords or len(word)==1:
            while word in Noun_Verb_words: Noun_Verb_words.remove(word)
    return Noun_Verb_words
```

```python
stopwords = []
docs_nouns_verbs = [Kr_preprocessing_noun_verb(doc,stopwords) for doc in total_docs]
```

```python
from collections import Counter

for i in range(20) :
  c = Counter(docs_nouns_verbs[i])
  print(c)
```

```python
#불용어 제거
stopwords=['했다','했습니다','하는','하며']
docs_nouns_verbs = [Kr_preprocessing_noun_verb(doc,stopwords) for doc in total_docs]
for i in range(20) :
  c = Counter(docs_nouns_verbs[i])
  print(c)
```

```python
documents_filtered_noun_verb = []
for doc in docs_nouns_verbs:
    document_filtered_noun_verb =' '.join(doc)
    documents_filtered_noun_verb.append(document_filtered_noun_verb)
documents_filtered_noun_verb[0]
```

# 3-2 TF 정보를 이용하여 벡터화하기

```python
from sklearn.feature_extraction.text import CountVectorizer # frequency based DTM
def tf_extractor(corpus): 
    # returns a frequency-based DTM
    vectorizer = CountVectorizer(min_df=1, ngram_range=(1,1))
    features = vectorizer.fit_transform(corpus) # transform texts to a frequency matrix
    return vectorizer, features 

vectorizer_tf_verb, DTM_tf_noun_verb = tf_extractor(documents_filtered_noun_verb) # 문서들을 TF matrix로 변환 
DTM_TF_noun_verb = np.array(DTM_tf_noun_verb.todense())
```
