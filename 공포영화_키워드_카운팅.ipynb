```python
ks = pd.read_csv("/content/drive/Shareddrives/✔DA 1차 플젝 2팀✔/review_crawling/곡성 크롤링.csv",encoding='cp949',header=None,names=["review"])
ks.head()
```

```python
ks["origin_text"] = ks["review"]
ks.head()
```

```python
search_keyword = ["오컬트","싸이코패스","연쇄살인","악마","좀비",
                  "미스터리","귀신","유령","심해","공포","심리적 공포",
                  "고어","잔인","스릴러","한국","외국","재난","여름"]



# for 문을 통해 해당 키워드가 있는지 여부를 True, False값으로 표시하도록 합니다.
# 키워드에 따라 컬럼을 새로 만듭니다.
for keyword in search_keyword:
    ks[keyword] = ks["review"].str.contains(keyword)
```

```python
ks.head()
```

```python
ks[search_keyword].sum().apply(f1).sort_values(ascending=False)
```
